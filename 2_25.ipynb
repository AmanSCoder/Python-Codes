```python
data2=pd.DataFrame({'key1':[1,2,45,5,67],
                    'key4':[4,5,6,7,8],
                    'key5':[3,45,5,6,6]})
```

```python
pd.merge(data1,data2) # this does the inner merge
# this will check for the column1, and whatever is common in both the columns
# it will bring all those rows into this new dataframe
```

```python
pd.merge(data1,data2,how='left') # this brings all the data of the first dataframe
# then it checks whether it is there in the second datafrome or not 
# if it is not there then it will simply put NaN in the place
```

```python
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='cross')
```

```python
pd.merge(data1,data2,how='outer',on='key1')
```

```python
# suppose now we want to change fare given in euro to the inr then 
# we will be using apply function( this is very much similar to the map fn)
```

```python
# now using the outersider function 

def euro_inr(x):
    return x*80

df['Fare_INR']=df['Fare'].apply(euro_inr)
```

```python
def cat_fare(x):
    if x<10:
        return 'cheap'
    elif x>=10 and x<20:
        return 'high'
    else:
        return 'high'
```

```python
df['car_fare']=df['Fare'].apply(cat_fare)
```
